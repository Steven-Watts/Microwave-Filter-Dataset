```python
import scipy.io
import numpy as np
import matplotlib.pyplot as plt
# Update the file path to the .mat file as needed
mat_data = scipy.io.loadmat(r"C:\Users\23548592\github\Microwave-Filter-Dataset\Testing Data\H-100.mat")
```

```python
# Extract all the data in the .mat file
grids = mat_data['dataset'][0,:]['grid']
solutions = mat_data['dataset'][0,:]['solution']
```

```python
selectionIdx = 1

grid = grids[selectionIdx,]
solution = solutions[selectionIdx,]
S = solution['S'][0,0]
F = solution['Freq'][0,0]
Z = solution['Zref'][0,0]

# Plot the binary array
plt.imshow(grid, cmap='summer', interpolation='nearest')
plt.title('32x32 Grid Data')
plt.show()
```

```python
S11 = S[0,0,:]
S12 = S[0,1,:]

# Plot the magnitude of S11 and S12
plt.plot(F,20*np.log10(np.abs(S11)), label='|S11|')
plt.plot(F,20*np.log10(np.abs(S12)), label='|S12|')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude (dB)')
plt.title('Magnitude of S11 and S12')
plt.legend()
plt.show()
```

```python
Z11 = Z[0,0,:]
Z12 = Z[0,1,:]

# Plot the magnitude of S11 and S12
plt.plot(F,Z11, label='Z11')
plt.plot(F,Z12, label='Z12')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude (dB)')
plt.title('Z11 and Z12')
plt.legend()
plt.show()
```
