```python
import scipy.io
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Update the file path to the .mat file as needed
mat_data = scipy.io.loadmat(r"C:\<...>\Microwave-Filter-Dataset\Testing Data\demo-10.mat")
```

```python
# Extract all the data in the .mat file
imageNames = mat_data['dataset'][0,:]['imageName']
topologies = mat_data['dataset'][0,:]['topology']
solutions = mat_data['dataset'][0,:]['solution']
```

```python
imagePath = r"C:\<...>\Microwave-Filter-Dataset\Testing Data\Images"
selectionIdx = 1

img_name = imageNames[selectionIdx][0]
img_full_path = imagePath + '\\' + img_name.split('\\')[-1]

img = mpimg.imread(img_full_path)
plt.imshow(img)
plt.axis('off')
plt.title('Selected Image')
plt.show()
```

```python
topology = topologies[selectionIdx,]
solution = solutions[selectionIdx,]
S = solution['S'][0,0]
F = solution['Freq'][0,0]
Z = solution['Zref'][0,0]

# Plot the binary array
plt.imshow(topology, cmap='summer', interpolation='nearest')
plt.title('32x32 Grid Data')
plt.show()
```

```python
S11 = S[0,0,:]
S12 = S[0,1,:]

# Plot the magnitude of S11 and S12
plt.plot(F,20*np.log10(np.abs(S11)), label='|S11|')
plt.plot(F,20*np.log10(np.abs(S12)), label='|S12|')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude (dB)')
plt.title('Magnitude of S11 and S12')
plt.legend()
plt.show()
```

```python
Z11 = Z[0,0,:]
Z12 = Z[0,1,:]

# Plot the magnitude of S11 and S12
plt.plot(F,Z11, label='Z11')
plt.plot(F,Z12, label='Z12')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude (dB)')
plt.title('Z11 and Z12')
plt.legend()
plt.show()
```
