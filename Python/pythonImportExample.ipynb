```python
import scipy.io
import numpy as np
import matplotlib.pyplot as plt
mat_data = scipy.io.loadmat(r"C:\Users\Steven Watts\LocalMasters\Microwave Filter Dataset\Microwave-Filter-Dataset\Training Data\1_Rotated_Noisy_X-500.mat")
```

```python
index = 5 # index of the dataset

# Extract the data
grid = mat_data['dataset'][0,index]['grid']
solutionData = mat_data['dataset'][0,index]['solution']
S = solutionData['S'][0,0]
F = solutionData['Freq'][0,0]
Z = solutionData['Zref'][0,0]
```

```python
# Plot the binary array
plt.imshow(grid, cmap='summer', interpolation='nearest')
plt.title('32x32 Grid Data')
plt.show()
```

```python
S11 = S[0,0,:]
S12 = S[0,1,:]

# Plot the magnitude of S11 and S12
plt.plot(F,20*np.log10(np.abs(S11)), label='|S11|')
plt.plot(F,20*np.log10(np.abs(S12)), label='|S12|')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude (dB)')
plt.title('Magnitude of S11 and S12')
plt.legend()
plt.show()
```

```python
Z11 = Z[0,0,:]
Z12 = Z[0,1,:]

# Plot the magnitude of S11 and S12
plt.plot(F,Z11, label='Z11')
plt.plot(F,Z12, label='Z12')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude (dB)')
plt.title('Z11 and Z12')
plt.legend()
plt.show()
```
